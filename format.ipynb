```python
import pandas as pd
from pathlib import Path
import glob
```

```python
csv_dir = Path('bu_courses')
csv_files = glob.glob(str(csv_dir / '*.csv'))

print(f"Found {len(csv_files)} CSV files:")
for file in csv_files:
    print(f"- {Path(file).name}")
```

```python
dfs = []

# Read each CSV and append to the list
for file in csv_files:
    try:
        df = pd.read_csv(file)
        dfs.append(df)
        print(f"Successfully read {Path(file).name} with {len(df)} rows")
    except Exception as e:
        print(f"Error reading {Path(file).name}: {str(e)}")

# Combine all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

# Display basic information about the combined dataset
print("\nCombined DataFrame Info:")
print(f"Total rows: {len(combined_df)}")
print(f"Total columns: {len(combined_df.columns)}")
print("\nColumns:")
for col in combined_df.columns:
    print(f"- {col}")
```

```python
combined_df_clean = combined_df.drop_duplicates()

# Check for and handle any missing values
print("Missing values in each column:")
print(combined_df_clean.isnull().sum())

# Display sample of the data
print("\nSample of combined data:")
display(combined_df_clean.head())

# Save the combined dataset
output_file = 'bu_all_courses.csv'
combined_df_clean.to_csv(output_file, index=False)
print(f"\nSaved combined dataset to {output_file}")
```

```python
combined_df_clean = combined_df_clean.replace('"', '', regex=True)
```

```python
COLLEGE_NAMES = {
    'BUA': 'Boston University Academy',
    'CAS': 'College of Arts & Sciences',
    'CASF': 'Faculty of Computing and Data Sciences',
    'CFA': 'College of Fine Arts',
    'CGS': 'College of General Studies',
    'COM': 'College of Communication',
    'ENG': 'College of Engineering',
    'EGS': 'College of Engineering and Graduate School of Arts & Sciences',
    'EOP': 'Center for English Language & Orientation Programs (CELOP)',
    'GMS': 'Graduate Medical Sciences',
    'GRS': 'Graduate School of Arts & Sciences',
    'HUB': 'Hub Office, General Education',
    'KHC': 'Kilachand Honors College',
    'LAW': 'School of Law',
    'MED': 'School of Medicine',
    'MET': 'Metropolitan College',
    'OTP': 'Officer Training Program',
    'PDP': 'Physical Development Program',
    'QST': 'Questrom School of Business',
    'SAR': 'Sargent College of Health & Rehabilitation Sciences',
    'SED': 'Wheelock College of Education & Human Development',
    'SDM': 'Henry M. Goldman School of Dental Medicine',
    'SHA': 'School of Hospitality Administration',
    'SPH': 'School of Public Health',
    'SSW': 'School of Social Work',
    'STH': 'School of Theology',
    'SUM': 'Summer Term Program',
    'UNI': 'The University Professors',
    'XRG': 'Cross Registration'
}
```

```python
import json

def parse_course_info(row):
    # Split course code from course name
    title_parts = row['title'].split(':', 1)
    full_course_code = title_parts[0].strip()
    course_name = title_parts[1].strip() if len(title_parts) > 1 else ''
    
    # Parse the full course code (e.g., "COM CM 180")
    code_parts = full_course_code.split()
    
    # Handle cases where we have 3 parts (division, department, number)
    if len(code_parts) >= 3:
        academic_unit = row['college']      # e.g., OTP (Previously called 'school')
        division = code_parts[0]            # e.g., COM (Previously called 'college')
        department = code_parts[1]          # e.g., CM
        number = code_parts[2]              # e.g., 180
    else:
        academic_unit = row['college']
        division = code_parts[0] if code_parts else ''
        department = ''
        number = code_parts[-1] if len(code_parts) > 1 else ''
    
    # Create simplified courseId (just department and number)
    course_id = f"{department}_{number}"
    
    return {
        'courseId': course_id,
        'university': 'Boston University',
        'academicUnit': academic_unit,
        'academicUnitName': COLLEGE_NAMES.get(academic_unit, academic_unit),
        'division': division,
        'divisionName': COLLEGE_NAMES.get(division, division),
        'department': department,
        'courseNumber': number,
        'fullCourseCode': full_course_code,
        'courseName': course_name,
        'description': row['description']
    }

# Convert the data
parsed_df = combined_df_clean.apply(parse_course_info, axis=1, result_type='expand')

```

```python
firebase_data = {
    'university': {
        'name': 'Boston University',
        'shortName': 'BU'
    },
    'courses': {
        f"{row['division']}_{row['courseId']}": {
            'university': row['university'],
            'academicUnit': row['academicUnit'],
            'academicUnitName': row['academicUnitName'],
            'division': row['division'],
            'divisionName': row['divisionName'],
            'department': row['department'],
            'courseNumber': row['courseNumber'],
            'fullCourseCode': row['fullCourseCode'],
            'courseName': row['courseName'],
            'description': row['description']
        }
        for _, row in parsed_df.iterrows()
    },
    'academicUnits': {
        unit.lower(): {
            'code': unit,
            'name': COLLEGE_NAMES.get(unit, unit),
            'divisions': list(parsed_df[parsed_df['academicUnit'] == unit]['division'].unique())
        }
        for unit in parsed_df['academicUnit'].unique()
    },
    'divisions': {
        div.lower(): {
            'code': div,
            'name': COLLEGE_NAMES.get(div, div),
            'departments': list(parsed_df[parsed_df['division'] == div]['department'].unique())
        }
        for div in parsed_df['division'].unique()
    }
}

output_file = 'bu_courses_firebase.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(firebase_data, f, ensure_ascii=False, indent=2)

print(f"\nProcessed {len(parsed_df)} courses")
print(f"Saved Firebase-structured data to {output_file}")
```
